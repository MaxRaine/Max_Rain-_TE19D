```python
def medelvärde(tal1,tal2):
    medel = (tal1+tal2)/2
    return medel

vikt_medel = medelvärde(60,62)

print("Min medelvikt är {vikt_medel}kg")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

print(f(5))
y = f(x)
x = np.linspace(-10,10)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("F(x) = x^2")
plt.grid()
plt.show()
```
